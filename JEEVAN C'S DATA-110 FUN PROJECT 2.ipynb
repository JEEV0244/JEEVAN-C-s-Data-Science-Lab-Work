```python
# This cell imports all required libraries for data handling and visualization
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# This cell creates an empty DataFrame to store stock price data
df = pd.DataFrame(columns=["Day", "Price"])
df
```

```python
# This cell takes input for number of days and starting stock price
days = int(input("Enter number of days to simulate: "))
initial_price = float(input("Enter starting stock price: "))
```

```python
# This cell initializes the price variable with starting value
price = initial_price
```

```python
# This cell simulates stock price changes for each day
for day in range(1, days + 1):
    change = random.uniform(-5, 5)  # random price fluctuation
    price = max(1, price + change) # ensure price doesn't go below 1

    new_row = pd.DataFrame([[day, round(price,2)]],
                           columns=["Day", "Price"])
    df = pd.concat([df, new_row], ignore_index=True)

df
```

```python
# This cell shows basic information about generated data
print("Total Days:", len(df))
print("Starting Price:", df["Price"].iloc[0])
print("Ending Price:", df["Price"].iloc[-1])
```

```python
# This cell creates a line graph of stock price over time
plt.plot(df["Day"], df["Price"], marker='o')
plt.title("Stock Price Trend")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
```

```python
# This cell takes investment amount and calculates number of shares bought
investment = float(input("Enter your investment amount: "))
buy_price = df["Price"].iloc[0]
shares = investment / buy_price

print("Shares Bought:", round(shares,2))
```

```python
# This cell calculates the value of investment for each day
df["Portfolio Value"] = df["Price"] * shares
df
```

```python
# This cell shows how your investment grows over time
plt.plot(df["Day"], df["Portfolio Value"], marker='o')
plt.title("Portfolio Growth")
plt.xlabel("Day")
plt.ylabel("Value")
plt.show()
```

```python
# This cell calculates final profit or loss clearly

final_value = df["Portfolio Value"].iloc[-1]
profit = final_value - investment

print("Final Value:", round(final_value, 2))

# Check whether profit or loss
if profit > 0:
    print("Actual Profit:", round(profit, 2))
elif profit < 0:
    print("Actual Loss:", round(abs(profit), 2))
else:
    print("No Profit No Loss")
```

```python
# This cell calculates daily percentage change in stock price
df["Change"] = df["Price"].pct_change()
df
```

```python
# This cell calculates volatility (risk measurement)
volatility = df["Change"].std()
print("Volatility:", volatility)
```

```python
# This cell classifies risk level based on volatility
if volatility < 0.02:
    risk = "Low"
elif volatility < 0.05:
    risk = "Medium"
else:
    risk = "High"

print("Risk Level:", risk)
```

```python
# This cell visualizes distribution of daily price changes
plt.hist(df["Change"].dropna())
plt.title("Price Change Distribution")
plt.xlabel("Change")
plt.ylabel("Frequency")
plt.show()
```

```python
# This cell finds highest and lowest stock prices
print("Highest Price:", df["Price"].max())
print("Lowest Price:", df["Price"].min())
```

```python
# This cell calculates moving average to analyze trend
df["Moving Avg"] = df["Price"].rolling(window=3).mean()
df
```

```python
# This cell compares actual price with moving average
plt.plot(df["Day"], df["Price"], label="Price")
plt.plot(df["Day"], df["Moving Avg"], label="Moving Avg")
plt.legend()
plt.title("Price vs Moving Average")
plt.show()
```

```python
# This cell gives suggestions based on profit and risk
print("🔍 Suggestion:")

if profit > 0 and risk == "Low":
    print("✅ Good investment, consider holding")
elif profit > 0 and risk == "Medium":
    print("👍 Moderate risk, monitor closely")
elif profit > 0 and risk == "High":
    print("⚠️ High risk, consider partial selling")
elif profit < 0:
    print("❌ Loss detected, reconsider strategy")
```

```python
# This cell prints final summary of the project
print("===== FINAL REPORT =====")
print("Initial Investment:", investment)
print("Final Value:", round(final_value,2))
print("Profit/Loss:", round(profit,2))
print("Risk Level:", risk)

if profit > 0:
    print("📈 Profit achieved")
else:
    print("📉 Loss occurred")
```
