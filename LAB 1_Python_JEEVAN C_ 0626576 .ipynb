**GIVEN PROBLEMS**

PROBLEM 1

```python
# Problem 1: Confusion Matrix Metrics
# Using Python as a Calculator

# Step 1: Assign given values to variables
TP = 50   # True Positives
FP = 8    # False Positives
TN = 32   # True Negatives
FN = 10   # False Negatives

# Step 2: Calculate total predictions
total_predictions = TP + FP + TN + FN

# Step 3: Calculate Accuracy
accuracy = (TP + TN) / total_predictions

# Step 4: Calculate Precision
precision = TP / (TP + FP)

# Step 5: Calculate Recall
recall = TP / (TP + FN)

# Step 6: Calculate F1 Score
f1_score = (2 * precision * recall) / (precision + recall)

# Step 7: Print results with clear labels
print("Total Predictions:", total_predictions)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

PROBLEM 2

```python
# Problem 2: R² (Coefficient of Determination)
# Using Python as a Calculator

# Step 1: Assign given values to variables
SSR = 420   # Sum of Squares due to Regression
SSE = 180   # Sum of Squares due to Error

# Step 2: Calculate Total Sum of Squares (SST)
SST = SSR + SSE

# Step 3: Calculate R-squared (R²)
R_squared = SSR / SST

# Step 4: Print results with clear labels
print("Sum of Squares due to Regression (SSR):", SSR)
print("Sum of Squares due to Error (SSE):", SSE)
print("Total Sum of Squares (SST):", SST)
print("R-squared (R²):", R_squared)

# Step 5: Interpretation
print("Interpretation:")
print("The model explains more than 70% variance:", R_squared > 0.70)
```

PROBLEM 3

```python
# Problem 3: Mean Squared Error (MSE)
# Using Python as a Calculator

# Step 1: Store squared errors in variables
e1 = 2
e2 = 4
e3 = 1
e4 = 9
e5 = 4

# Step 2: Calculate the sum of squared errors
sum_squared_errors = e1 + e2 + e3 + e4 + e5

# Step 3: Store the number of observations
n = 5

# Step 4: Calculate Mean Squared Error (MSE)
MSE = sum_squared_errors / n

# Step 5: Print the result with a clear label
print("Sum of Squared Errors:", sum_squared_errors)
print("Number of Observations:", n)
print("Mean Squared Error (MSE):", MSE)
```

PROBLEM 4

```python
# Problem 4: Cook’s Distance
# Using Python as a Calculator

# Step 1: Assign given values to variables
e_squared = 9      # Squared residual (e_i^2)
MSE = 3            # Mean Squared Error
p = 3              # Number of predictors
h_ii = 0.2         # Leverage value
n = 25             # Sample size

# Step 2: Calculate Cook's Distance
# Formula: D_i = (e_i^2 / (p * MSE)) * (h_ii / (1 - h_ii)^2)

cooks_distance = (e_squared / (p * MSE)) * (h_ii / ((1 - h_ii) ** 2))

# Step 3: Calculate threshold value (4/n)
threshold = 4 / n

# Step 4: Print results with clear labels
print("Cook's Distance:", cooks_distance)
print("Threshold (4/n):", threshold)

# Step 5: Decision on influence
print("Is the observation influential?:", cooks_distance > threshold)
```

PROBLEM 5

```python
# Problem 5: Simple Linear Regression Prediction
# Using Python as a Calculator

# Step 1: Assign given values to variables
x = 12          # Rainfall value
slope = 2.5     # Slope of the regression line
intercept = 10  # Intercept of the regression line

# Step 2: Calculate predicted value of y
y = (slope * x) + intercept

# Step 3: Print the predicted output
print("Given value of x (Rainfall):", x)
print("Predicted value of y (Crop Yield):", y)

# Step 4: Explanation of slope and intercept
print("Explanation:")
print("Slope (2.5) indicates the change in crop yield for each unit increase in rainfall.")
print("Intercept (10) represents the crop yield when rainfall is zero.")
```
