```python
import pandas as pd
from datetime import datetime
import random
import string

# Create empty DataFrame
expenses_df = pd.DataFrame(columns=["ID", "Date", "Category", "Amount"])

# Function to generate ID
def generate_id():
    return ''.join(random.choices(string.ascii_uppercase + string.digits, k=7))

# Add multiple entries
n = int(input("How many expenses to add: "))

for i in range(n):
    print(f"\nEntry {i+1}")

    expense_id = generate_id()
    date = datetime.today().strftime('%Y-%m-%d')

    category = input("Category: ").capitalize()   # ✅ Fix: make consistent
    amount = float(input("Amount: "))

    new_row = pd.DataFrame([[expense_id, date, category, amount]],
                           columns=["ID", "Date", "Category", "Amount"])

    expenses_df = pd.concat([expenses_df, new_row], ignore_index=True)

# ✅ Show Data
print("\nAll Expenses:")
print(expenses_df)

# ✅ FINAL TOTAL (IMPORTANT)
total = expenses_df["Amount"].sum()
print("\nTotal Expense:", total)
```

```python
# Remove invalid entries
expenses_df = expenses_df[expenses_df["Amount"] > 0]

# Convert date to datetime
expenses_df["Date"] = pd.to_datetime(expenses_df["Date"])

expenses_df
```

```python
print("Average Expense:", expenses_df["Amount"].mean())
print("Highest Expense:", expenses_df["Amount"].max())
```

```python
# Basic statistics
total = expenses_df["Amount"].sum()
avg = expenses_df["Amount"].mean()
max_exp = expenses_df["Amount"].max()

print("Total Expense:", total)
print("Average Expense:", round(avg, 2))
print("Highest Expense:", max_exp)
```

```python
# Category analysis
category_totals = expenses_df.groupby("Category")["Amount"].sum()

print(category_totals)

top_category = category_totals.idxmax()
print("\nHighest spending category:", top_category)
```

```python
# Daily trend
daily_expense = expenses_df.groupby("Date")["Amount"].sum()

print(daily_expense)
```

```python
# Pie chart
category_totals.plot.pie(autopct='%1.1f%%', title="Expense Distribution")
plt.ylabel("")
plt.show()
```

```python
# Bar chart
category_totals.plot.bar(title="Category-wise Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```

```python
# Insights
print("\n🔍 Insights:")

if avg > 500:
    print("- Your average spending is high ⚠️ ")
else:
    print("- Your spending is under control ✅")

print(f"- You spend the most on {top_category}")
print(f"- Total number of entries: {len(expenses_df)}")
```

```python
category_total = expenses_df.groupby("Category")["Amount"].sum()

print("Category-wise Total:")
print(category_total)
```

```python
highest_category = category_total.idxmax()
highest_value = category_total.max()

print("Highest Spending Category:", highest_category)
print("Amount:", highest_value)
```

```python
print("===== FINAL SUMMARY =====")
print("Total Expense:", expenses_df["Amount"].sum())
print("Average Expense:", round(expenses_df["Amount"].mean(), 2))
print("Highest Expense:", expenses_df["Amount"].max())
print("Lowest Expense:", expenses_df["Amount"].min())
print("Top Category:", highest_category)
```

```python
print("🔍 Insights:")

if expenses_df["Amount"].sum() > 5000:
    print("- You are spending a lot ⚠️")
else:
    print("- Your spending is under control ✅")

print("- Most money spent on:", highest_category)
print("- Total entries:", len(expenses_df))
```
