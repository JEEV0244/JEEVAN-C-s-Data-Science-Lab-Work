```python
# =========================================================
# FACEBOOK DATA ANALYSIS
# =========================================================
# Name: JEEVAN C
# Subject: DATA-110 Introduction to Python Programming for Machine Learning
# =========================================================

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make graphs beautiful
plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

print("All libraries imported successfully")

```

In this step, required Python libraries were imported for data analysis and visualization.

-Pandas is used for data manipulation and analysis

-NumPy is used for numerical operations

-Matplotlib and Seaborn are used for data visualization

Graph styles were applied to improve the clarity and presentation of visualizations.

```python
# =========================================================
# STEP 2: DATA ACCESS
# Reading TXT, CSV and Excel files
# =========================================================

# TXT file (auto detect separator)
df_txt = pd.read_csv("Facebook Friends.txt", sep=None, engine='python', encoding='latin-1')

# CSV file
df_csv = pd.read_csv("Facebook Friends.csv")

# Excel file
df_excel = pd.read_excel("Facebook Friends.xlsx")

print("All files loaded successfully")
```

In this step, the Facebook dataset was loaded using three different file formats: TXT, CSV, and Excel.

Separate dataframes were created:

-df_txt for TXT file

-df_csv for CSV file

-df_excel for Excel file

The purpose of this step is to check whether file format affects data loading in Python.

All files were successfully loaded into Python using pandas functions.

```python
# =========================================================
# Compare all file types
# =========================================================

print("\nShape of TXT file:", df_txt.shape)
print("Shape of CSV file:", df_csv.shape)
print("Shape of Excel file:", df_excel.shape)

print("\nColumn names in Excel file:")
print(df_excel.columns)

print("\nChecking if dataframes are equal")
print("TXT vs CSV:", df_txt.equals(df_csv))
print("CSV vs Excel:", df_csv.equals(df_excel))
```

The shape and column names of each dataframe were compared to ensure consistency.

The .equals() function was used to compare datasets from different file formats.

Conclusion:
The three file formats show differences in shape and structure when loaded into Python.
The Excel file contains complete and properly formatted data, while TXT and CSV files show structural differences.
Therefore, Excel file is more suitable and reliable for further analysis.

```python
# =========================================================
# Using Excel dataframe for analysis
# =========================================================

df = df_excel.copy()

print("\nFirst 5 rows")
display(df.head())

print("\nLast 5 rows")
display(df.tail())

print("\nColumn names")
print(df.columns)

print("\nData types")
print(df.dtypes)

print("\nDataset information")
df.info()
```

Basic data examination functions were used to understand the structure of the dataset.

-head() displays first 5 rows

-tail() displays last 5 rows

-columns shows column names

-dtypes shows data types of each column

-info() provides complete dataset information

These functions help in understanding dataset size, structure, and variable types before analysis.

```python
# =========================================================
# Missing values
# =========================================================

print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nTotal missing values:")
print(df.isnull().sum().sum())
```

Missing values were checked using isnull() function.

Missing data can affect analysis and visualization accuracy.
The total number of missing values in each column and entire dataset was calculated.

Conclusion:
If missing values are zero, dataset is clean and ready for analysis.
If present, data cleaning would be required before analysis.

```python
# =========================================================
# Statistical summary
# =========================================================

print("\nStatistical Summary:")
display(df.describe())
```

The describe() function provides statistical summary of numerical columns.

It includes:

-Count of values

-Mean (average)

-Standard deviation

-Minimum value

-25%, 50%, 75% quartiles

-Maximum value

How to read the table:
Mean shows average values of each column.Standard deviation shows variation in data.Minimum and maximum show range of data.

These statistics help to understand distribution and behavior of numerical data.

```python
# =========================================================
# Histograms for numerical data
# =========================================================

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(8,5))
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}", fontsize=14)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Histograms were plotted for numerical columns to visualize data distribution.

Histogram helps to:

-Understand frequency of values

-Identify data spread

-Detect skewness and patterns

Visualization makes data easier to understand compared to raw numbers.

```python
# =========================================================
# FREQUENCY PLOTS FOR CATEGORICAL DATA
# =========================================================

cat_cols = df.select_dtypes(include=['object']).columns

print("Categorical columns are:", cat_cols)

if len(cat_cols) == 0:
    print("No categorical columns available in dataset for frequency plot.")
    print("All columns are numerical. Hence histogram is more suitable.")
else:
    for col in cat_cols:
        plt.figure(figsize=(9,5))
        sns.countplot(x=df[col], palette="viridis")
        plt.title(f"Frequency Distribution of {col}", fontsize=15, fontweight='bold')
        plt.xticks(rotation=45)
        plt.show()

```

Frequency plots (count plots) were created for categorical columns.

These plots show how many times each category appears in dataset.
It helps to identify most common categories and understand categorical distribution.

If no categorical columns are present, frequency plots will not be generated.

```python
# =========================================================
# STEP: CORRELATION HEATMAP
# =========================================================
# Heatmap shows relationship between numerical variables
# =========================================================

plt.figure(figsize=(12,7))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            fmt=".2f")

plt.title("Correlation Heatmap of Numerical Variables", fontsize=16, fontweight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

Correlation heatmap shows relationship between numerical variables.

-Value close to +1 indicates strong positive correlation

-Value close to -1 indicates negative correlation

-Value near 0 indicates weak correlation

This visualization helps to identify relationships between variables and understand how one variable affects another.

```python
# =========================================================
# EXTRA VISUAL 1: BOXPLOT
# =========================================================

numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot of {col}", fontsize=15)
    plt.xlabel(col)
    plt.show()
```

```python
# =========================================================
# EXTRA VISUAL 2: PAIRPLOT
# =========================================================

numeric_df = df.select_dtypes(include='number')

if numeric_df.shape[1] >= 2:
    sns.pairplot(numeric_df)
    plt.suptitle("Pairplot of Numerical Variables", y=1.02)
    plt.show()
else:
    print("Not enough numerical columns for pairplot")
```

```python
# =========================================================
# EXTRA VISUAL 2: BAR CHART FOR NUMERICAL COLUMNS
# =========================================================

numeric_cols = df.select_dtypes(include='number').columns

if len(numeric_cols) > 0:
    df[numeric_cols].mean().plot(kind='bar', figsize=(10,5))
    plt.title("Average Values of Numerical Columns", fontsize=15)
    plt.xlabel("Columns")
    plt.ylabel("Average Value")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No numerical columns available for bar chart")
```
